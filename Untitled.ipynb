```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
def zero_pad(X, pad):
    X_pad = np.pad(X, ((0,0),(pad, pad),(pad, pad),(0,0)), 'constant') #np.pad 用法参见https://blog.csdn.net/zenghaitao0128/article/details/78713663
    return X_pad
```

```python
test_for_zero_pad = np.random.randn(4,3,3,2)
test_for_zero_pad_pad = zero_pad(test_for_zero_pad, 2)
print(test_for_zero_pad)
print("////")

print(test_for_zero_pad[0,:,:,0])

print("////")

test_for_zero_pad_pad[0,:,:,0]
```

```python
fig, axarr = plt.subplots(1,2)
axarr[0].set_title('x')
axarr[0].imshow(test_for_zero_pad[0,:,:,0])
axarr[1].set_title('x_pad')
axarr[1].imshow(test_for_zero_pad_pad[0,:,:,0])
plt.show()
```
